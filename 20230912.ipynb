```python
## 첫번째: 제일 잘한 과목의 점수 best_score
## 두번째: 제일 잘한 과목명 쓰기
## 세전째: 잘한 과목명 순서대로 쓰기
'math --> korea --> science --> english'
# person.iloc[0,1:5].max()
# person.iloc[:,1:5].apply(np.max,1)
bs = np.argmax(person.iloc[0,1:5]) #제일 잘한 과목의 위치
person.columns[1:5][bs] ##어디(bs)를 찾았으면 컬럼에 그 위치를 집어넣어 컬럼명을 찾음
def find_subject(x):
    bs = np.argmax(x)
    return person.columns[1:5][bs]
```

```python
person['best_subject'] = person.iloc[:,1:5].apply(find_subject,1)
```

```python
person.rename(columns={'best_score':'best_subject'},inplace=True)
```

```python
person['best_score']=person.iloc[:,1:5].apply(np.max,1)
```

```python
'-->'.join(person.columns[1:5][np.argsort(person.iloc[0,1:5])[::-1]])
def subject_order(x):
    #subject = np.array(['korea','english','math',science'])
    #person.columns[1:5]대신 subject를 넣어도 무방함
    return '-->'.join(person.columns[1:5][np.argsort(x)[::-1]])
```

```python
'-->'.join(person.columns[1:5][np.argsort(person.iloc[1,1:5])[::-1]])
```

```python
person['grade_order']=person.iloc[:,1:5].apply(subject_order, 1)
```

```python
#goupby : split (컬럼), apply, combine
group_df = pd.DataFrame(columns = ['best_subject','korea','english'])
for i in person.best_subject.unique():
    tmp_df = person.loc[person.best_subject==i,:].sum()[["korea","english"]]
    tmp_df = pd.DataFrame(
        {
            'best_subject':[i],
            'korea':tmp_df.value[[0]],
            'english':tmp_df.value[[1]]
        }
    )
    group_df = pd.concat([group_df,tmp_df])
group_df.set_index('best_subject')

person.groupby(['best-subject']).sum()[['korea','english']]
```
